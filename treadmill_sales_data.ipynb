```python
def find_probabilities(gender, marginal_probability = False):
    if gender != 'Male' and gender != 'Female':
        return 'Invalid entry'
    df1 = pd.crosstab(index = df['Gender'], columns = df['Product'])
    p_kp281 = df1['KP281'][gender] / df1.loc[gender].sum()
    p_kp481 = df1['KP481'][gender] / df1.loc[gender].sum()
    p_kp781 = df1['KP781'][gender] / df1.loc[gender].sum()
    
    if marginal_probability:
        print(f'Marginal Probability of Male:- {df1.loc["Male"].sum()/len(df1):.2f}')
        print(f'Marginal Probability of Female:- {df1.loc["Female"].sum()/len(df1):.2f}')
        
    print(f'P(KP281/{gender}):- {p_kp281: .2f}')
    print(f'P(KP481/{gender}):- {p_kp481:.2f}')
    print(f'P(KP781)/{gender}):- {p_kp781:.2f}')
    
find_probabilities('Male', True)
find_probabilities('Female')
```

```python
def find_probabilities(MaritalStatus, marginal_probability = False):
    if MaritalStatus != 'Partnered' and MaritalStatus != 'Single':
        return 'Invalid Entry'
    df1 = pd.crosstab(index = df['MaritalStatus'], columns = df['Product'])
    p_kp281 = df1['KP281'][MaritalStatus] / df1.loc[MaritalStatus].sum()
    p_kp481 = df1['KP481'][MaritalStatus] / df1.loc[MaritalStatus].sum()
    p_kp781 = df1['KP781'][MaritalStatus] / df1.loc[MaritalStatus].sum()
    
    if marginal_probability:
        print(f'Probability of Partnered:- {df1.loc["Partnered"].sum()/len(df1): .2f}')
        print(f'Probability of Single:- {df1.loc["Single"].sum()/len(df1): .2f}')
        
    print(f'P(KP281/{MaritalStatus}):- {p_kp281 :.2f}')
    print(f'P(KP481/{MaritalStatus}):- {p_kp481 :.2f}')
    print(f'P(KP781/{MaritalStatus}):- {p_kp781 :.2f}')
    
find_probabilities('Partnered', True)
find_probabilities('Single')
```
